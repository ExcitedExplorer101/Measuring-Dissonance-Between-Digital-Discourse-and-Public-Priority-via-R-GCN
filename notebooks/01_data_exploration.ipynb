#### Loading Dataset

```python
data_path = "../data/usc-x-24-us-election/part_1/"

files = [
    "may_july_chunk_1.csv.gz",
    "may_july_chunk_2.csv.gz",
    "may_july_chunk_3.csv.gz",
    "may_july_chunk_4.csv.gz",
    "may_july_chunk_5.csv.gz"
]
```

#### Load All 5 Chunks

```python
dfs = []
for f in files:
    print("Loading:", f)
    temp = pd.read_csv(data_path + f, compression='gzip')
    dfs.append(temp)

df = pd.concat(dfs, ignore_index=True)

print("Total rows:", len(df))
df.head()
```

#### Dataset Stats

```python
stats = {
    "Total Tweets": len(df),
    "Unique Users": df['user'].nunique(),
    "Unique Conversations": df['conversationId'].nunique(),
    "Total Retweets": df['retweetedTweet'].notna().sum(),
    "Total Replies": df['in_reply_to_user_id_str'].notna().sum(),
    "Total Quotes": df['quotedTweet'].notna().sum(),
    "Total Tweets with Hashtags": df['hashtags'].notna().sum(),
    "Total Tweets with Mentions": df['mentionedUsers'].notna().sum(),
    "Total Tweets with Links": df['links'].notna().sum(),
    "Languages Used": df['lang'].nunique()
}

stats_df = pd.DataFrame(list(stats.items()), columns=['Metric', 'Value'])
stats_df
```

```python
import plotly.express as px

fig = px.bar(
    stats_df,
    x='Metric',
    y='Value',
    title='Dataset Statistics',
    text='Value'
)

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

#### Tweet Interaction Types

```python
interaction_counts = {
    "Hashtags": df['hashtags'].notna().sum(),
    "Mentions": df['mentionedUsers'].notna().sum(),
    "Links": df['links'].notna().sum(),
    "Replies": df['in_reply_to_user_id_str'].notna().sum(),
    "Retweets": df['retweetedTweet'].notna().sum(),
    "Quotes": df['quotedTweet'].notna().sum()
}

interaction_df = pd.DataFrame(list(interaction_counts.items()), columns=['Type', 'Count'])

fig = px.pie(interaction_df, names='Type', values='Count', hole =0.3, title='Tweet Interaction Types')
fig.update_traces(
    textinfo='label+percent+value'
)
fig.show()
```

#### Engagement Distribution

```python
engagement_cols = ['replyCount', 'retweetCount', 'likeCount', 'quoteCount']

engagement_df = df[engagement_cols].fillna(0)
engagement_sum = engagement_df.sum()

import plotly.express as px

fig = px.pie(
    names=engagement_sum.index,
    values=engagement_sum.values,
    hole = 0.3,
    title='Engagement Distribution'
)

fig.update_traces(
    textinfo='label+percent+value'
)

fig.show()
```

#### Top Hashtags

```python
df['hashtags'].dropna().head(10)
```

```python
import ast
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

def extract_hashtag_text(x):
    if pd.isna(x):
        return []
    try:
        tags = ast.literal_eval(x)
        return [t['text'] for t in tags if 'text' in t]
    except:
        return []

df['hashtags_list'] = df['hashtags'].apply(extract_hashtag_text)

all_tags = [tag for sublist in df['hashtags_list'] for tag in sublist]

# tag_counts = pd.Series(all_tags).value_counts().head(20).reset_index()
tag_counts = pd.Series(all_tags).value_counts().reset_index()
tag_counts.columns = ['hashtag', 'count']

# Take top 20
tag_counts = tag_counts.head(20)

# Sort so largest appears on top in horizontal bar
tag_counts = tag_counts.sort_values(by='count', ascending=True)

fig = go.Figure(
    go.Bar(
        x=tag_counts['count'],
        y=tag_counts['hashtag'],
        orientation='h'
    )
)

fig.update_layout(
    title='Top 20 Hashtags',
    xaxis_title='Count',
    yaxis_title='Hashtag',
    yaxis=dict(
        tickmode='array',
        tickvals=tag_counts['hashtag'],
        ticktext=tag_counts['hashtag']
    )
)

fig.show()

```
